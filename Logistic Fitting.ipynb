```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
n = 20
xs = [30,40,50,60,70,75,80,90,100,120]
xs = np.array([[x] * n for x in xs]).flatten()
ys = []
def add_to(ys, n_yes, n=n):
    ys.extend([1] * n_yes)
    ys.extend([0] * (n - n_yes))
add_to(ys, 0) # 30 (1)
add_to(ys, 0) # 40 (2)
add_to(ys, 0) # 50 (3)
add_to(ys, 0) # 60 (4)
add_to(ys, 5) # 70 (5)
add_to(ys, 10) # 75 (6)
add_to(ys, 15) # 80 (7)
add_to(ys, 20) # 90 (8)
add_to(ys, 20) # 100 (9)
add_to(ys, 20) # 120 (10)
fig, ax = plt.subplots(dpi=200, layout='constrained')
sns.regplot(x=xs,y=ys, logistic=True, scatter_kws={'alpha': 0.05, 's': 100})
plt.yticks(np.linspace(0,1,6), (np.linspace(0,1,6)*100).astype(int).astype(str))
plt.ylabel('Percent Fooled', fontsize=16)
plt.xlabel('Size of Squares', fontsize=16)
plt.show()
```
